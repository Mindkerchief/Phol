```python
# Load 150 epochs training log from results.csv
yolov3m_150_results = pd.read_csv('results/yolov3m-150.csv')
yolov5m_150_results = pd.read_csv('results/yolov5m-150.csv')
yolov6m_150_results = pd.read_csv('results/yolov6m-150.csv')
yolov8m_150_results = pd.read_csv('results/yolov8m-150.csv')
yolov9m_150_results = pd.read_csv('results/yolov9m-150.csv')
yolov10m_150_results = pd.read_csv('results/yolov10m-150.csv')
yolov11m_150_results = pd.read_csv('results/yolov11m-151.csv')
yolov12m_150_results = pd.read_csv('results/yolov12m-150.csv')

# Load 250 epochs training log from results.csv
yolov11m_250_results = pd.read_csv('results/yolov11m-250.csv')
yolov12m_250_results = pd.read_csv('results/yolov12m-250.csv')
yolov12n_250_results = pd.read_csv('results/yolov12n-250.csv')

# Extract mAP values
yolov3m_150_mAP50 = yolov3m_150_results["metrics/mAP50(B)"].tolist()
yolov3m_150_mAP50_95 = yolov3m_150_results["metrics/mAP50-95(B)"].tolist()

yolov5m_150_mAP50 = yolov5m_150_results["metrics/mAP50(B)"].tolist()
yolov5m_150_mAP50_95 = yolov5m_150_results["metrics/mAP50-95(B)"].tolist()

yolov6m_150_mAP50 = yolov6m_150_results["metrics/mAP50(B)"].tolist()
yolov6m_150_mAP50_95 = yolov6m_150_results["metrics/mAP50-95(B)"].tolist()

yolov8m_150_mAP50 = yolov8m_150_results["metrics/mAP50(B)"].tolist()
yolov8m_150_mAP50_95 = yolov8m_150_results["metrics/mAP50-95(B)"].tolist()

yolov9m_150_mAP50 = yolov9m_150_results["metrics/mAP50(B)"].tolist()
yolov9m_150_mAP50_95 = yolov9m_150_results["metrics/mAP50-95(B)"].tolist()

yolov10m_150_mAP50 = yolov10m_150_results["metrics/mAP50(B)"].tolist()
yolov10m_150_mAP50_95 = yolov10m_150_results["metrics/mAP50-95(B)"].tolist()

yolov11m_150_mAP50 = yolov11m_150_results["metrics/mAP50(B)"].tolist()
yolov11m_150_mAP50_95 = yolov11m_150_results["metrics/mAP50-95(B)"].tolist()

yolov12m_150_mAP50 = yolov12m_150_results["metrics/mAP50(B)"].tolist()
yolov12m_150_mAP50_95 = yolov12m_150_results["metrics/mAP50-95(B)"].tolist()

yolov11m_250_mAP50 = yolov11m_250_results["metrics/mAP50(B)"].tolist()
yolov11m_250_mAP50_95 = yolov11m_250_results["metrics/mAP50-95(B)"].tolist()

yolov12m_250_mAP50 = yolov12m_250_results["metrics/mAP50(B)"].tolist()
yolov12m_250_mAP50_95 = yolov12m_250_results["metrics/mAP50-95(B)"].tolist()

yolov12n_250_mAP50 = yolov12n_250_results["metrics/mAP50(B)"].tolist()
yolov12n_250_mAP50_95 = yolov12n_250_results["metrics/mAP50-95(B)"].tolist()

# Plot the results
plt.figure(figsize=(12, 5))
plt.plot(yolov3m_150_mAP50, label="YOLOv3-M")
plt.plot(yolov5m_150_mAP50, label="YOLOv5-M")
plt.plot(yolov6m_150_mAP50, label="YOLOv6-M")
plt.plot(yolov8m_150_mAP50, label="YOLOv8-M")
plt.plot(yolov9m_150_mAP50, label="YOLOv9-M")
plt.plot(yolov10m_150_mAP50, label="YOLOv10-M")
plt.plot(yolov11m_150_mAP50, label="YOLOv11-M")
plt.plot(yolov12m_150_mAP50, label="YOLOv12-M")
# plt.plot(yolov11m_250_mAP50, label="YOLOv11-M")
# plt.plot(yolov12m_250_mAP50, label="YOLOv12-M")
# plt.plot(yolov12n_250_mAP50, label="YOLOv12-N")
plt.xlabel("Epochs")
plt.ylabel("mAP")
plt.title("mAP@50 Progression Over 150 Epochs")
plt.legend()
plt.grid()
plt.show()
```

```python
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')

# Apply moving average to the mAP@50 values
window_size = 10
yolov3_mAP50_smooth = moving_average(yolov3m_150_mAP50, window_size)
yolov5_mAP50_smooth = moving_average(yolov5m_150_mAP50, window_size)
yolov6_mAP50_smooth = moving_average(yolov6m_150_mAP50, window_size)
yolov8_mAP50_smooth = moving_average(yolov8m_150_mAP50, window_size)
yolov9_mAP50_smooth = moving_average(yolov9m_150_mAP50, window_size)
yolov10_mAP50_smooth = moving_average(yolov10m_150_mAP50, window_size)
yolov11_mAP50_smooth = moving_average(yolov11m_150_mAP50, window_size)
yolov12_mAP50_smooth = moving_average(yolov12m_150_mAP50, window_size)

# Plot the smoothed results
plt.figure(figsize=(12, 5))
plt.plot(yolov3_mAP50_smooth, label="YOLOv3m (Smoothed)")
plt.plot(yolov5_mAP50_smooth, label="YOLOv5m (Smoothed)")
plt.plot(yolov6_mAP50_smooth, label="YOLOv6m (Smoothed)")
plt.plot(yolov8_mAP50_smooth, label="YOLOv8m (Smoothed)")
plt.plot(yolov9_mAP50_smooth, label="YOLOv9m (Smoothed)")
plt.plot(yolov10_mAP50_smooth, label="YOLOv10m (Smoothed)")
plt.plot(yolov11_mAP50_smooth, label="YOLOv11m (Smoothed)")
plt.plot(yolov12_mAP50_smooth, label="YOLOv12m (Smoothed)")
plt.xlabel("Epochs")
plt.ylabel("mAP")
plt.title("Smoothed mAP@50 Progression Over 150 Epochs")
plt.legend()
plt.grid()
plt.show()
```

```python
yolov3_box_loss = yolov3m_150_results["train/dfl_loss"].tolist()
yolov5_box_loss = yolov5m_150_results["train/dfl_loss"].tolist()
yolov6_box_loss = yolov6m_150_results["train/dfl_loss"].tolist()
yolov8_box_loss = yolov8m_150_results["train/dfl_loss"].tolist()
yolov9_box_loss = yolov9m_150_results["train/dfl_loss"].tolist()
yolov10_box_loss = yolov10m_150_results["train/dfl_loss"].tolist()
yolov11_box_loss = yolov11m_150_results["train/dfl_loss"].tolist()
yolov12_box_loss = yolov12m_150_results["train/dfl_loss"].tolist()

# Plot the smoothed results
plt.figure(figsize=(12, 5))
plt.plot(yolov3_box_loss, label="YOLOv3m")
plt.plot(yolov5_box_loss, label="YOLOv5m")
plt.plot(yolov6_box_loss, label="YOLOv6m")
plt.plot(yolov8_box_loss, label="YOLOv8m")
plt.plot(yolov9_box_loss, label="YOLOv9m")
plt.plot(yolov10_box_loss, label="YOLOv10m")
plt.plot(yolov11_box_loss, label="YOLOv11m")
plt.plot(yolov12_box_loss, label="YOLOv12m")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Box Loss Progression Over 150 Epochs")
plt.legend()
plt.grid()
plt.show()
```

```python
def get_best_metrics(results):
    best_epoch = results.loc[results['metrics/mAP50(B)'].idxmax()]
    return {
        'mAP@50': best_epoch['metrics/mAP50(B)'],
        'mAP@50-95': best_epoch['metrics/mAP50-95(B)']
    }

# Get metrics for each model
metrics = {
    'YOLOv6': get_best_metrics(yolov6m_150_results),
    'YOLOv5': get_best_metrics(yolov5m_150_results),
    'YOLOv9': get_best_metrics(yolov9m_150_results),
    'YOLOv10': get_best_metrics(yolov10m_150_results),
    'YOLOv12': get_best_metrics(yolov12m_150_results),
    'YOLOv3': get_best_metrics(yolov3m_150_results),
    'YOLOv8': get_best_metrics(yolov8m_150_results),
    'YOLOv11': get_best_metrics(yolov11m_150_results),
}

# Prepare data for plotting
categories = ['mAP@50', 'mAP@50-95']
models = list(metrics.keys())
values = {category: [metrics[model][category] for model in models] for category in categories}

# Plot grouped bar chart
x = np.arange(len(models))
width = 0.2
fig, ax = plt.subplots(figsize=(10, 5))

# Plot bars for each metric
for i, category in enumerate(categories):
    ax.bar(x + i * width, values[category], width, label=category)

ax.set_xlabel('Models')
ax.set_ylabel('Precision')
ax.set_title('Comparison of mAP YOLO Models')
ax.set_xticks(x + width * (len(categories) - 1) / 2)
ax.set_xticklabels(models)
ax.legend(loc='lower right')
plt.tight_layout()

# Increase the width of the bars
for bar_group in ax.containers:
    for bar in bar_group:
        bar.set_width(bar.get_width() * 2.5)

plt.bar_label(ax.containers[0], fmt='%.3f')
plt.bar_label(ax.containers[1], fmt='%.3f')
plt.grid(axis='y', linestyle='--')
plt.show()
```
